# Histogram

# A histogram is a way of representing the distribution of a dataset.
# It groups continuous (or large discrete) data into intervals (bins) and shows how many data points fall into each bin.

# X-axis → bins (ranges of values)

# Y-axis → frequency (count of data points in that bin)

```python
plt.figure(figsize=(15,6))
sns.distplot(DF['Age']) #histogram
plt.show()
```

# For the x-axis, we are giving Age and the histogram is by default include kernel density estimate (kde).

# KDE is the curved line along with the bins or the edges of the frequency of the Ages.

# If you want to remove the Kernel density estimate (kde) then use kde = False.



```python
plt.figure(figsize=(15,6))
sns.distplot(DF['Age'],kde=False)
plt.show()
```

```python
plt.figure(figsize=(15,7))
sns.distplot(DF['CC'],kde=False)
plt.show()
```

# After that, you got frequency as the y-axis and the age of the car as the x-axis.

# If you want to organize all the different intervals or bins, you can use the bins parameter on the distplot function.

# Let’s use bins = 5 on the distplot function. It will organize your bins into five bins or intervals.

```python
plt.figure(figsize=(15,6))
sns.distplot(DF['Price'],kde=False,bins=10)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create synthetic data
np.random.seed(42)
data1 = np.random.normal(loc=50, scale=15, size=500)   # Normal distribution centered at 50
data2 = np.random.normal(loc=65, scale=10, size=500)   # Another dataset



# Create figure for histogram cheat sheet
fig, axes = plt.subplots(3, 2, figsize=(12, 14))
fig.suptitle("Histogram Use Cases Cheat Sheet", fontsize=16, fontweight="bold")



# 1. Simple histogram
axes[0,0].hist(data1, bins=10, color="skyblue", edgecolor="black", alpha=0.7)
axes[0,0].set_title("1. Simple Histogram")



# 2. Different bin sizes
axes[0,1].hist(data1, bins=30, color="orange", edgecolor="black", alpha=0.7)
axes[0,1].set_title("2. Different Bin Sizes (30 bins)")



# 3. Normalized (density) histogram
axes[1,0].hist(data1, bins=20, density=True, color="green", edgecolor="black", alpha=0.6)
axes[1,0].set_title("3. Normalized Histogram (Density=True)")



# 4. Multiple histograms (overlay)
axes[1,1].hist(data1, bins=20, alpha=0.5, label="Dataset 1", color="blue")
axes[1,1].hist(data2, bins=20, alpha=0.5, label="Dataset 2", color="red")
axes[1,1].legend()
axes[1,1].set_title("4. Overlay of Multiple Histograms")



# 5. Cumulative histogram
axes[2,0].hist(data1, bins=20, cumulative=True, color="purple", edgecolor="black", alpha=0.7)
axes[2,0].set_title("5. Cumulative Histogram")



# 6. Horizontal histogram
axes[2,1].hist(data1, bins=15, orientation="horizontal", color="teal", edgecolor="black", alpha=0.7)
axes[2,1].set_title("6. Horizontal Histogram")



plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# Sample dataset
np.random.seed(42)
df = pd.DataFrame({
    "Region": np.random.choice(["North", "South", "East", "West"], 20),
    "Product": np.random.choice(["A", "B", "C", "D", "E"], 20),
    "Department": np.random.choice(["HR", "IT", "Sales"], 20),
    "Month": np.random.choice(["Jan", "Feb", "Mar", "Apr"], 20),
    "Sales": np.random.randint(50, 200, 20),
    "Target": np.random.randint(100, 180, 20),
    "Profit": np.random.randint(-50, 100, 20),
    "Male": np.random.randint(10, 50, 20),
    "Female": np.random.randint(10, 50, 20),
    "Revenue": np.random.randint(200, 500, 20)
})


# Create subplots (4x2 grid for 8 examples)
fig, axes = plt.subplots(4, 2, figsize=(14, 18))
axes = axes.flatten()


# 1. Compare Categories
df.groupby("Region")["Sales"].sum().plot(kind="bar", ax=axes[0], color="skyblue", edgecolor="black")
axes[0].set_title("1. Compare Categories (Sales by Region)")



# 2. Ranking (Sorted Barh)
df.groupby("Product")["Revenue"].sum().sort_values().plot(kind="barh", ax=axes[1], color="orange")
axes[1].set_title("2. Ranking (Revenue by Product)")



# 3. Composition (Stacked Bar)
df.groupby("Department")[["Male", "Female"]].sum().plot(kind="bar", stacked=True, ax=axes[2])
axes[2].set_title("3. Composition (Gender by Department)")


""

# 4. Grouped Bars
df.groupby("Month")[["Sales", "Target"]].sum().plot(kind="bar", ax=axes[3])
axes[3].set_title("4. Grouped Bars (Sales vs Target by Month)")


# 5. Deviation Bars (Profit/Loss)
profits = df.groupby("Month")["Profit"].sum()
profits.plot(kind="bar", ax=axes[4], color=["g" if p > 0 else "r" for p in profits])
axes[4].axhline(0, color="black", linewidth=1)
axes[4].set_title("5. Deviation (Profit/Loss by Month)")


# 6. 100% Stacked Bar (Proportion)
shares = df.groupby("Region")[["Male", "Female"]].sum()
shares.div(shares.sum(axis=1), axis=0).plot(kind="bar", stacked=True, ax=axes[5])
axes[5].set_title("6. Proportion (100% Stacked by Region)")



#7. Mixed Plot (Bar + Line)
sales = df.groupby("Month")["Sales"].sum()
target = df.groupby("Month")["Target"].sum()
ax = axes[6]
sales.plot(kind="bar", color="lightblue", ax=ax, label="Sales")
target.plot(kind="line", color="red", marker="o", ax=ax, label="Target")
ax.set_title("7. Mixed Plot (Sales vs Target)")
ax.legend()



# 8. Top-N Categories
df.groupby("Product")["Revenue"].sum().nlargest(5).plot(kind="bar", ax=axes[7], color="teal")
axes[7].set_title("8. Top-N (Top 5 Products by Revenue)")


plt.tight_layout()
plt.show()
```

```python
✅ This script will generate 8 bar plot variations in a single cheat sheet:

# Compare Categories

# Ranking (sorted bars)

# Stacked composition

# Grouped bars

# Deviation (profit/loss)

# 100% stacked (proportion)

# Mixed bar + line plot

# Top-N categories
```

# Box and Whiskers Plot:


# Box and whiskers plots are used for analyzing the detailed distribution of a dataset.

# Let’s plot Box and whiskers plot of the Price column of the dataset to visually interpret the “five-number summary”.

# ########### Five Number Summary includes ##########

# Minimum (smallest value excluding outliers)

# First Quartile (Q1, 25th percentile)

# Median (Q2, 50th percentile)

# Third Quartile (Q3, 75th percentile)

# Maximum (largest value excluding outliers)

```python
          Outliers (●)
   ┌─────────────────────────────
   |        ┌─────────────┐
 Min ──┐    |             |    ┌── Max
       └────┤   Median    ├────┘
            |             |
            └─────────────┘
            Q1            Q3
```

```python
# Box = Interquartile Range (IQR = Q3 − Q1)

# Line inside box = Median (Q2)

# Whiskers = Extend to min/max values within 1.5×IQR

# Dots = Outliers beyond whiskers
```

```python
DF = pd.read_csv("ToyotaCorolla.csv",encoding = "ISO-8859-1", engine='python')
print(DF.head(10))

plt.figure(figsize=(17,6))
sns.boxplot(y=DF["Price"])
plt.show()
```

# Minimum:  The minimum value of the dataset excluding the outliers

# Maximum:  The maximum value of the dataset excluding the outliers

# First quartile/Q1 (25th percentile): The median value between the smallest number and the median of the dataset

# Median (50th percentile):       The median of the Dataset.

# Third Quartile/Q3 (75th percentile):  The median value between the median of the dataset and the highest value of the dataset

# Here, IQR is interquartile range:
IQR = Q3-Q1                                           
Anything above the whiskers is called Outliers. Outlier or extreme values lies above 1.5 times the median values.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Sample dataset
np.random.seed(42)
df = pd.DataFrame({
    "Maths": np.random.normal(70, 10, 100),
    "Science": np.random.normal(65, 15, 100),
    "English": np.random.normal(75, 8, 100),
    "Group": np.random.choice(["A", "B"], 100),
    "Gender": np.random.choice(["Male", "Female"], 100)
})


# Create figure
fig, axes = plt.subplots(4, 2, figsize=(14, 18))
axes = axes.flatten()


# 1. Basic Boxplot
df[["Maths", "Science", "English"]].plot(kind="box", ax=axes[0])
axes[0].set_title("1. Basic Boxplot")



# 2. Horizontal Boxplot
df[["Maths", "Science", "English"]].plot(kind="box", vert=False, ax=axes[1])
axes[1].set_title("2. Horizontal Boxplot")



# 3. Boxplot by Group
sns.boxplot(x="Group", y="Maths", data=df, ax=axes[2])
axes[2].set_title("3. Group-wise Boxplot (Maths by Group)")



# 4. Multiple Groups Comparison
sns.boxplot(data=df[["Maths","Science","English"]], ax=axes[3])
axes[3].set_title("4. Multiple Variables Boxplot")



# 5. Boxplot with Strip Plot (Overlay)
sns.boxplot(x="Group", y="Science", data=df, ax=axes[4])
sns.stripplot(x="Group", y="Science", data=df, color="red", alpha=0.5, ax=axes[4])
axes[4].set_title("5. Boxplot + Strip Plot")



# 6. Boxplot with Swarm Plot (Overlay)
sns.boxplot(x="Gender", y="English", data=df, ax=axes[5])
sns.swarmplot(x="Gender", y="English", data=df, color="black", size=3, ax=axes[5])
axes[5].set_title("6. Boxplot + Swarm Plot")



# 7. Customized Boxplot (Colors & Notches)
sns.boxplot(data=df[["Maths","Science","English"]], notch=True,
            palette="Set2", ax=axes[6])
axes[6].set_title("7. Notched & Colored Boxplot")



# 8. Outlier Detection Example
sns.boxplot(y=df["Maths"], ax=axes[7])
axes[7].set_title("8. Outliers Highlighted (Maths)")



plt.tight_layout()
plt.show()
```

```python
✅ This cheat sheet will show 8 variations:

# Basic boxplot

# Horizontal boxplot

# Group-wise boxplot

# Multi-variable boxplot

# Boxplot + strip plot

# Boxplot + swarm plot

# Notched & colored boxplot

# Outlier visualization
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("iris")


# Create subplots grid
fig, axes = plt.subplots(2, 2, figsize=(14, 12))


# Basic pairplot
sns.pairplot(df)
plt.suptitle("📊 Visual Cheat Sheet for Pairplots", fontsize=18, y=1.02)



# Variation 1: Hue for categories
sns.pairplot(df, hue="species")
plt.suptitle("🎨 Pairplot with Hue", fontsize=16, y=1.02)



# Variation 2: Regression lines
sns.pairplot(df, kind="reg", hue="species")
plt.suptitle("📈 Pairplot with Regression Lines", fontsize=16, y=1.02)



# Variation 3: KDE on diagonal + Scatter
sns.pairplot(df, diag_kind="kde", hue="species")
plt.suptitle("🔥 Pairplot with KDE Diagonals", fontsize=16, y=1.02)



# Variation 4: Corner plot (no redundancy)
sns.pairplot(df, hue="species", corner=True)
plt.suptitle("🟦 Corner Pairplot", fontsize=16, y=1.02)


plt.show()
```

# ✅ Code: Pairplot Cheat Sheet with Real-World Dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load real-world dataset (Restaurant tips)
df = sns.load_dataset("tips")

# 1. Basic pairplot
sns.pairplot(df)
plt.suptitle("📊 Basic Pairplot - Tips Dataset", fontsize=16, y=1.02)
plt.show()

# 2. Hue: Differentiate by Gender
sns.pairplot(df, hue="sex")
plt.suptitle("🎨 Pairplot with Hue (Sex)", fontsize=16, y=1.02)
plt.show()

# 3. Regression Lines
sns.pairplot(df, kind="reg", hue="sex")
plt.suptitle("📈 Pairplot with Regression Lines", fontsize=16, y=1.02)
plt.show()

# 4. KDE Diagonal
sns.pairplot(df, diag_kind="kde", hue="sex")
plt.suptitle("🔥 Pairplot with KDE Diagonals", fontsize=16, y=1.02)
plt.show()

# 5. Corner Plot
sns.pairplot(df, hue="sex", corner=True)
plt.suptitle("🟦 Corner Pairplot (Sex)", fontsize=16, y=1.02)
plt.show()
```
